Aqui o APP e Aplicativo são o mesmo canal de vendas, então usar o comando abaixo para substituir "APP" por "Aplicativo"

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

Aqui tem a coluna "estado" e a coluna "Preço" com dados nulos.

```python
vendas[vendas["estado"].isnull()]
```

```python
#vendas_sem_nulos = vendas.dropna()
#vendas_sem_nulos.info()
# para excluir as célular com nulos - nessa função, todas as linhas onde existem nulos são excluidas tb
```

```python
#premissa de negócio: onde eu não tenho estado (nulo), considerar MS.
vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```python
media_preco = vendas["Preço"].mean()
media_preco = float(media_preco)
vendas["Preço"].fillna(media_preco, inplace=True)
```

```python
vendas[["Preço_com_frete", "Preço"]].describe()
```

```python
#preço não pode ser maior do que preço com frete
#criar uma regra para eliminar essas células com erro
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#departamento mais vendido
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#média de preço com frete por departamento
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
#Quantidade de vendas por mês
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correto["mes"]= vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/DNC - DATA PREPARATION/[PD] 05 - Materiais/cliente_varejo.xlsx")
cliente.head()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.head()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
#média de renda para cada tipo de canal de vendas
agg_idcanal_venda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_venda.head()
```

```python
#média de idade de clientes por bandeira
agg_bandeira_idade = round(vendas_cliente.groupby("bandeira")["idade"].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_bandeira_idade.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_bandeira_idade["bandeira"], agg_bandeira_idade["idade"], color = "green")
plt.ylabel("Média de Idade")
plt.title("Idade média por bandeira", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_venda ["idcanalvenda"], agg_idcanal_venda["renda"], color = "blue")
plt.ylabel("Renda Média")
plt.title("Renda por Canal de Venda", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
```

```python
vendas_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(20, 12))

plt.xlabel("Data da Venda")
plt.ylabel("Quantidade de Vendas")
x = vendas_por_data["Data"]
y = vendas_por_data["idcompra"]

plt.plot(x, y, color = "green", linewidth = "2")
plt.show()
```

```python
#média de preço com frete por departamento
agg_dep_preço = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dep_preço
```

```python
fig = plt.figure(figsize=(30,20))

plt.bar(agg_dep_preço ["Nome_Departamento"], agg_dep_preço["Preço_com_frete"], color = "red")
plt.ylabel("Preço com frete")
plt.xlabel("Departamento")
plt.title("Preço com frete por Departamento", loc = "center", fontsize = 15, fontweight = "bold", color = "black");
```
